We are interested in finding an optimal sequence of queries to ask a user to obtain a more precise idea of their utility. Optimal is meant in the sense that the sequence should provide the most information for the least query cost. 

## The Domain

The negotiation domain $\Omega$ is a "multiple split the pie" domain. There are $n$ pies that need to be shared between 2 negotiators $N_1$ and $N_2$. We let then $\Omega = [0,1]^n$, where each outcome $\omega\in\Omega$ corresponds to the amount of each pie given to $N_1$. For example $\omega = (1,1\cdots,1,0)$ means $N_1$ gets all the first $n-1$ pies and the last pie goes to $N_2$. $\omega = (0.5,0.5,\cdots,0.5)$ means each negotiator gets half of each pie. ## The Model

Assume we are the agent that is representing $N_1$. It is safe to assume that $N_1$ wants more of any given pie. However, the importance attached to getting a particular pie may differ. For this reason, we model the utility of $N_1$ with the following linear addtive utiltiy functions $\mathcal{U}$:
    
$$\mathcal{U}=\left\{\omega\mapsto\sum_{i=1}^n \theta_i\cdot \omega_i | \theta_i\geq 0 \forall i, \sum_{i=1}^n \theta_i = 1\right\}$$

We assume that the true utility $u^*$ of $N_1$ is in $\in\mathcal{U}$. It follows that we are looking for the corresponding vector of parameters $\theta^*\in\Theta:=\{\theta\in\mathbb[0.1]^n | \sum_{i=1}^n\theta_i=1\}$.


```python
n = 3 # Number of pies
true_theta = np.random.dirichlet([1 for i in range(n)]) # Generate the true parameter uniformly at random

'''Computes the utility of an outcome.'''
def utility(outcome, theta):
    return sum([outcome[i]*theta[i] for i in range(n)])

print(true_theta)
```

## Querying

It is possible to query the user for information on $u^*$ against a cost. The queries considered are pairwise comparisons of outcomes, namely: 

$$\mathcal{Q}=\{(\omega\geq \omega')? | \omega,\omega\in\Omega\}$$

where $(\omega\geq \omega')?$ is to be understood as: "Is $u^*(\omega)\geq u^*(\omega')$ ?" or "Do you prefer outcome $\omega$ over outcome $\omega'$?" The possible answers are simply 1 (Yes) if indeed $\omega$ is prefered, or 0 (No) otherwise. We also assume constant cost of queries. Namely, for each $q\in\mathcal{Q}, c(q) = c$

```python
query_cost = 0.01 # Constant cost initialization
cum_cost = 0 # Cummulative cost

def query(outcome_1, outcome_2):
    # Update cumulative cost
    global cum_cost
    global query_cost
    cum_cost += query_cost
    
    # Compute the true utilities of the outcomes and compare them
    if(utility(outcome_1, true_theta) >= utility(outcome_2,true_theta)):
        return 1
    return 0

#EXAMPLE#
outcome_1 = [0,0.5,1]
outcome_2 = [1,0.5,0]
print("theta =", true_theta)
print("u((0,0.5,1)) = ", utility(outcome_1,true_theta))
print("u((1,0.5,0)) = ", utility(outcome_2,true_theta))

if(query(outcome_1,outcome_2)):
    print("(0,0.5,1) is prefered over (1,0.5,0)")
else:
    print("(1,0.5,0) is prefered over (0,0.5,1)")

#Reset cum_cost after example
cum_cost = 0
```

## Querying to Reduce Uncertainty

Before querying the user, we do not have much information on the true parameter $\theta^*$ except that it is in $\Theta$ which in our case is equal to the $n-1$-dimensional simplex $\Delta^{n-1}$: 
$$\Delta^{n-1}=\{(\theta_1,\cdots,\theta_n) | \sum_{i=1}^n\theta_i= 1, \theta_i\geq 0 \forall i\}$$
Below is a visualization of $\Theta$ for $n=3$.

The absence of prior information on $\theta^*$ can be formalized with a uniform belief over $\Theta$. Namely, every $\theta\in\Theta$ is potentially equal to $\theta^*$. Because $\Theta$ is equal to the standard simplex, this belief can be characterised by a Dirichlet distribution with parameter $\alpha = (1)^n\in\mathbb{R}^n$. Let then $\Pi = \text{Dir}(1,\cdots,1)$.

### The Effect of Querying

Consider a query $q\in\mathcal{Q}$. What happens if we ask $q$ and receive a correct answer $a(\theta^*,q)$? Well, the answer will correspond to an inequality $u^*(\omega_q)\geq u^*(\omega_q')$. This inequality can in turns be carefully rewritten as $\sum_{i=1}^n x_{q,i} \theta^*_i\leq 0$ where $x_{q,i}\in\mathbb{R}$. This defines a half-space, and we thus conclude that $\theta^*$ must be in the intersection of $\Theta$ and this halfspace. This intersection is defined as the set $\Theta_{a^*(q)}$:

$$\Theta_{a^*(q)}:=\{\theta\in\Theta | a(\theta,q) = a(\theta^*,q)\} = \left\{\theta\in\Theta |\sum_{i=1}^n x_{q,i} \theta_i\leq 0\right\}$$

Observe that $\Theta_{a^*(q)}$ is a convex polytope given in its $H-$description ($H$: halfspaces). Furthermore, asking more queries and receiving their answers allows us to define for a set $Q\subseteq \mathcal{Q}$, the convex polytope $\Theta_{a^*(Q)}$: The intersection of $\Theta$ with the halfspaces corresponding to answered queries in $Q$.
The $V-$description ($V$: vertex) is more convenient to use for representation. Below is a method for switching to $V-$description.


```python
'''
INPUT: a list of answered queries
An answered query is a 3-tuple of this shape (o_1, o_2, query(o_1,o_2))

OUTPUT: The vertex representation of the polytope defined by the answered queries. 
'''
def vertex_rep(answered_queries): 
    m = len(answered_queries)
    
    # Initialize the input 
    A = [[0]*n for i in range((n+2)+m)]
    b = [0]*((n+2)+m)
    
    # Start with the basic n+2 conditions: -theta_i <= 0 and sum(theta_i) <= 1 and sum(-theta_i) <= -1
    for i in range(n):
        A[i][i] = -1
        A[n][i] = 1
        A[n+1][i] = -1
    b[n] = 1
    b[n+1] = -1
    
    # Continue by converting the answered queries into a valid input for compute_polytope_vertices
    for i in range(m):
        o_1 = answered_queries[i][0]
        o_2 = answered_queries[i][1]
        answer = answered_queries[i][2]
        
        if answer == 1: # Taking care of the inequality sign of u^*(o_1) >= u^*(o_2)
            A[i+n+2] = [o_2[j]-o_1[j] for j in range(n)] 
        else:
            A[i+n+2] = [o_1[j]-o_2[j] for j in range(n)]
    
    # Finish by computing the vertices
    return compute_polytope_vertices(np.array(A),np.array(b))
```

Continuing with our $n=3$ dimensional example, consider the query $q=((1,0,0)\geq (0,1,0))?$ Here is what $\Theta_{a^*(q)}$ looks like:

```python
#Ask q and record the answer
a_q = ([1,0,0],[0,1,0],query([1,0,0],[0,1,0]))

#Compute Theta_{a^*(q)}
V = [tuple(v.tolist()) for v in vertex_rep([a_q])]

fig = plt.figure()

ax = Axes3D(fig)
vertices = [V]
ax.add_collection3d(Poly3DCollection(vertices))
for v in V:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")

ax.view_init(elev=10., azim=100)

plt.show()
print (V)
```

### Reducing uncertainty

We have seen that asking a query $q$ has the effect to cut our parameter space $\Theta$ into 2 parts. Receiving the answer $a^*(q)$ tells us in which part $\theta^*$ is in. We called this part $\Theta_{a^*(q)}$. Belief wise, our belief was essentially updated from $\Pi(\theta)$ to $\Pi(\theta|(q, a^*(q))$:

$$\Pi(\theta|(q, a^*(q)) \propto \Pi(\theta, (q,a^*(q)) = \begin{cases} \Pi(\theta) \text{ if } \theta\in\Theta_{a^*(q)}\\ 0 \text{ otherwise} 
\end{cases}$$

Following an information theoretical approach, it follows that when answered, a query reduces uncertainty by discarding an amount of probability mass from our belief. This mass corresponds to $\Pi(\Theta\backslash \Theta_{a^*(q)})$. Prior to asking any query, we represent each query's information potential by the following quantity:

$$I(q):= -\log{\dfrac{\Pi(\Theta\backslash \Theta_{a^*(q)})}{\Pi(\Theta)}}$$

Note that we don't know $a^*(q)$. Besides, we have no reason to assume a query has more chances of being answered yes or no. So rather than assuming a certain distribution for $a^*(q)$, we redefine $I(q)$ to be the $\textbf{worst case}$ information potential:

$$I(q):= \min_{a\in\{0,1\}} -\log{\dfrac{\Pi(\Theta\backslash \Theta_{(q,a)})}{\Pi(\Theta)}}$$

Remark that $I(q)=0$ means that $q$ does not give us any information. Namely, it doesn't actually cut $\Theta$ into 2 parts because there is a way to answer $q$ that leads to $\Theta_{a^*(q)}=\Theta$. A trivial example of such a query would be $(1,1,1)\geq (0,0,0)?$ Its interpreation is that the user obviously prefers all the 3 pies than none of the pies.  

To reduce uncertainty as much as possible, we would like to ask $q$ which maximizes $I(q)$ without incurring to much cost. Namely, find $q_1$ s.t.:

$$q_1 \in \text{argmax}_{q\in\mathcal{Q}} I(q) - c(q)$$

Since we want to continue asking queries after $q_1$, we will need to use an adjusted definition for $I(q)$ to subsequently find the optimal query to ask. For any given subset of answered queries $a^*(Q)$, let:
$$I(a^*(Q),q):= \min_{a\in\{0,1\}} -\log{\dfrac{\Pi\left(\Theta_{a^*(Q)}\backslash \Theta_{a^*(Q)\cup\{(q,a)\}}\right)}{\Pi\left(\Theta_{a^*(Q)}\right)}}$$

#### Objective

Find $q_1 \in \text{argmax}_{q\in\mathcal{Q}} I(q) - c(q)$, and $q_2,\cdots,q_T$ such that:

$$q_{t} \in \text{argmax}_{q\in\mathcal{Q}} I(a^*(\{q_1,\cdots,q_{t-1}\}),q) - c(q)$$

AND

$$I(a^*(\{q_1,\cdots,q_{t-1}\}),q_t) - c(q) > 0$$

## Computing $I(q)$  

From now on, we are exclusively concerned with the example $n=3$.
Since $\Pi$ is Dir$(1,1,1)$, it is a uniform distribution over the $2-simplex$ $\Theta$. Hence for $S\subseteq \Theta$,  $\Pi(S)$ is proportional to the surface area of $S$. It follows that  

$$I(q) =\min_{a\in\{0,1\}} -\log{\dfrac{\text{Area}(\Theta\backslash \Theta_{(q,a)})}{\text{Area}(\Theta)}}$$

To compute $I(q)$, we must be able to compute the surface area of a convex polytope. We can use the ConvexHull library of scipy to do that.

```python
'''
INPUT: answered queries
OUTPUT: the surface area of the convex polytope they define.
'''
def area(answered_queries):
    vertices = [v.tolist() for v in vertex_rep(answered_queries)]
    
    #Check that we actually have a valid polytope with enough vertices
    if(len(vertices)<=2):
        return 0.001
   
    #Add an extra coplanar point to have a non degenerate convex hull
    extra_v = [sum([v[i] for v in vertices])/len(vertices) for i in range(3)]
    vertices.append(extra_v)
    V = np.asarray(vertices)
    
    #Use 'QJ' to add a tiny bit of volume to the hull. The area is then approximately area/2 
    #because the polytope has 2 superposed identical face.
    hull = ConvexHull(V, qhull_options ='QJ')
    return hull.area/2


#EXAMPLE: The simplex itself is an equilateral triangle with side length sqrt(2). Its area is approximately 0.86
print("Initial THETA has area ", area([]), " in the 3 dimensional space")
    
```

We are now ready to compute the worst case information potential $I(q)$ of a query $q$.

```python
'''
Helper detecting uncondtionaly useless queries, e.g. (1,1,1) >= (0,0,1)?
'''

def useless(q):
    d_1 = q[0][0] - q[1][0]
    d_2 = q[0][1] - q[1][1]
    d_3 = q[0][2] - q[1][2]
    if((d_1 >= 0 and d_2 >= 0 and d_3 >= 0) or (d_1 <= 0 and d_2 <= 0 and d_3 <= 0)):
        return True
    return False

'''
INPUT: a list of already answered queries, a query q
OUTPUT: The worst case information potential of q
'''

def info_potential(answered_queries, q):
   
    if(useless(q)):
        return 0
    
    base_area = area(answered_queries)
    
    # Answered queries if q were to be answered no
    q_0 = (q[0],q[1],0)
    Q_0 = answered_queries.copy()
    Q_0.append(q_0)
    
    
    # Answered queries if q were to be answered yes
    q_1 = (q[0],q[1],1)
    Q_1 = answered_queries.copy()
    Q_1.append(q_1)
    
    
    # Largest area is worst case area
    area_0 = area(Q_0)
    area_1 = area(Q_1)
    worst_area = max(area_0,area_1)
    i_q = -math.log(worst_area/base_area)
    
    return i_q


#EXAMPLE

#Ask 
q = ([1,1,0],[1,0.5,1])
print(info_potential([],q))
```

## Querying Algorithms

### Approximate optimal querying

This is an approximation of what should optimally be done. As we know, the query space $\mathcal{Q}$ is infinite. Rather than analytically finding the information maximizing queries, we consider the discretization of $\mathcal{Q}$:

$$\mathcal{Q}_m = \left\{\left((\omega_1,\cdots,\omega_n)\geq (\omega'_1,\cdots \omega'_n)\right)? | \forall i, \omega_i,\omega'_i\in \left\{0,\dfrac{1}{m},\cdots,1\right\}\right\}$$

We then iterate over $\mathcal{Q}_m$ to find the optimal next query, and repeat the process.


```python
'''
We first do a little helper step to compute the discretized outcome space O
'''

def discretize(m):
    base_values = [float(i)/m for i in range(m+1)]
    space = [base_values.copy() for i in range(n)]
    O = itertools.product(*space)
    return list(O)

'''
INPUT: A list of answered queries over a domain with any number of pies. An integer m to discretize the space

OUTPUT: A tuple of: (Approximation of the next best query to ask, its information potential)
'''

def approximate_opt_query(answered_queries, m):
    O = discretize(m)
    x = 0
    best_query = []
    
    # Iterate over possible q and compute I(q)
    for o_1 in O:
        for o_2 in O:
            q = (list(o_1),list(o_2))
            temp = info_potential(answered_queries,q)
            if(temp > x):
                x = temp
                best_query = list(q)
    
    if(len(best_query)>0): #There is indeed a best query to ask
        print("Best query is: ",best_query, ", with info potential: ",x)
        return (best_query,x)
    
    print("No query adds info ")
    return -1
        
#######EXAMPLE########

'''
Let's say we choose m=2. Start at the begining.
'''

m=2
answered_queries = []
V_0 = [tuple(l.tolist()) for l in vertex_rep(answered_queries)]

#Visualization
print("Polytope -- 0:")
fig0 = plt.figure()
ax = Axes3D(fig0)
ax.add_collection3d(Poly3DCollection([V_0]))

for v in V_0:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")
ax.view_init(elev=10., azim=100)

plt.show()

#Find the first best query
query_1 = approximate_opt_query(answered_queries, m)[0]
#Ask the query and add to the list of answers.
a_1 = query(query_1[0],query_1[1])
answered_queries.append((query_1[0],query_1[1],a_1))

#Visualization
V_1 = [tuple(l.tolist()) for l in vertex_rep(answered_queries)]
print(V_1)
print("Polytope -- 1:")
fig1 = plt.figure()
ax = Axes3D(fig1)
ax.add_collection3d(Poly3DCollection([V_1]))

for v in V_1:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")
ax.view_init(elev=10., azim=100)

plt.show()

#Repeat 2 more times
query_2 = approximate_opt_query(answered_queries, m)[0]
a_2 = query(query_2[0],query_2[1])
answered_queries.append((query_2[0],query_2[1],a_2))

#Visualization
V_2 = [tuple(l.tolist()) for l in vertex_rep(answered_queries)]
print("Polytope -- 2:")
fig2 = plt.figure()
ax = Axes3D(fig2)
ax.add_collection3d(Poly3DCollection([V_2]))

for v in V_2:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")
ax.view_init(elev=10., azim=100)

plt.show()


query_3 = approximate_opt_query(answered_queries, m)[0]
a_3 = query(query_3[0],query_3[1])
answered_queries.append((query_3[0],query_3[1],a_3))

#Visualization
V_3 = [tuple(l.tolist()) for l in vertex_rep(answered_queries)]
print("Polytope -- 3:")
fig3 = plt.figure()
ax = Axes3D(fig3)
ax.add_collection3d(Poly3DCollection([V_3]))

for v in V_3:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")
ax.view_init(elev=10., azim=100)

plt.show()

query_4 = approximate_opt_query(answered_queries, m)[0]
a_4 = query(query_4[0],query_4[1])
answered_queries.append((query_4[0],query_4[1],a_4))

#Visualization
V_4 = [tuple(l.tolist()) for l in vertex_rep(answered_queries)]
print("Polytope -- 4:")
fig4 = plt.figure()
ax = Axes3D(fig4)
ax.add_collection3d(Poly3DCollection([V_4]))

for v in V_4:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")
ax.view_init(elev=10., azim=100)

plt.show()

query_5 = approximate_opt_query(answered_queries, m)[0]
a_5 = query(query_5[0],query_5[1])
answered_queries.append((query_5[0],query_5[1],a_5))
#Visualization
V_5 = [tuple(l.tolist()) for l in vertex_rep(answered_queries)]
print("Polytope -- 5:")
fig5 = plt.figure()
ax = Axes3D(fig5)
ax.add_collection3d(Poly3DCollection([V_5]))

for v in V_5:
    text = str(round(v[0],2))
    for i in range(len(v)-1):
        text+= ", " + str(round(v[i+1],2))
    ax.text(v[0],v[1],v[2], text, zdir=(0, 0, 0))
    
ax.set_xlabel("theta 1")
ax.set_ylabel("theta 2")
ax.set_zlabel("theta 3")
ax.view_init(elev=10., azim=100)

plt.show()
```
